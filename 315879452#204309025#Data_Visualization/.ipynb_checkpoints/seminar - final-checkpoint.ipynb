#### Imports 

```python
import matplotlib.pyplot as plt
import seaborn as sns
from enum import Enum,auto
from tkinter import *
from PIL import ImageTk, Image
from tkinter import ttk
import tkinter
from tkcalendar import Calendar
import time
from pandastable import Table, TableModel
import pandas as pd
import numpy as np
import time
from tkinter.filedialog import askopenfile
from tkinter.ttk import *
from matplotlib.backends.backend_tkagg import FigureCanvasTkAgg
from PIL import Image, ImageTk
import numpy as np
from urllib.request import urlopen
```

#### CODE

```python
class Type(Enum):
    STUDENTS = 0
    COURSES = 1
    LECRURER = 2
    TAKE = 3

class Program:
    search_list = ['Students','Lecturers','Courses','Take']
    PAD = 10
    def __init__(self):
        super().__init__()
        self.root = Tk()
        self.students = pd.read_csv('Students.csv')
        self.students = self.students.set_index(self.students.columns[0])
        self.lecturers = pd.read_csv('Lecturers.csv')
        self.lecturers.set_index(self.lecturers.columns[0],drop=True)
        self.courses = pd.read_csv('Courses.csv').set_index('ID',drop=True)
        self.courses.set_index(self.courses.columns[0],drop=True)
        self.take = pd.read_csv('Take.csv')
        self.title_name = tkinter.Label(self.root,text="MENU".upper())
        self.title_name.pack()
        self.menu_page()  
            
    def _make_main_frame(self):
        self.main_frm = ttk.Frame(self.root)
        self.main_frm.pack(padx=self.PAD,pady=self.PAD,fill=BOTH,expand=True)
    
    def _make_entry(self):
        ent = ttk.Entry(self.main_frm,justify='right')
        ent.pack()
        return ent
        
    def _make_label(self,text):
        label = tkinter.Label(self.main_frm,text=text,font=('Helvetica', 10, 'bold'))
        label.pack(pady=15)
        return label
        
    def menu_page(self):
        self.clear()
        self.root.geometry("800x600")
        frame_btn = tkinter.Frame(self.root)
        frame_btn.pack(anchor="c")
        frame_btn.place(anchor="c", relx=.5, rely=.5)
        self.title('MENU')
 
        self.bn_search = tkinter.Button(master=frame_btn,text="Show",command= self.search_page, height = 5, width = 10,font=('Helvetica', 10, 'bold')).pack(pady=10)
        self.bn_update = tkinter.Button(master=frame_btn,text="Update",command= self.update_page, height = 5, width = 10,font=('Helvetica', 10, 'bold')).pack(pady=10)
        self.bn_cancel = tkinter.Button(master=frame_btn,text="Exit",command=self.root.destroy, height = 5, width = 10,font=('Helvetica', 10, 'bold')).pack(pady=10)
        self.main()
    
    def update_page(self):
        self.root.geometry("1000x1000")
        self.clear()
        self.title('Update')
        self._make_main_frame()
        value_inside = tkinter.StringVar(self.main_frm)
        value_inside.set("Select an Option")
        question_menu = tkinter.OptionMenu(self.main_frm, value_inside, *Program.search_list,command=self.add_info)
        question_menu.pack()
        frame = Frame(master=self.root).pack()
        self.back = Button(master=frame,text="Back",command=self.menu_page).pack(pady=10)
        self.bn_cancel = Button(master=frame,text="Exit",command=self.root.destroy).pack(pady=10)
        
    def add_info(self,value_inside):
        self.clean(1)
        if value_inside == 'Students':
            self.df = self.students
            self.build_info('Students.csv',value_inside)
        elif  value_inside == 'Lecturers':
            self.df = self.lecturers
            self.build_info('Lecturers.csv',value_inside)
        elif  value_inside == 'Courses':
            self.df = self.courses
            self.build_info('Courses.csv',value_inside)
        else:
            self.df = self.take
            self.build_info('Take.csv',value_inside)   
    def build_info(self,path,value_inside): 
        a = []
        for i in self.df.columns:
                l = Label(master=self.main_frm , text= f'{i}',font=('Helvetica', 10, 'bold'))
                l.pack(padx= 2)
                b = Entry(master=self.main_frm , text= f'{i}',font=('Helvetica', 10, 'bold'))
                b.pack(pady = (0,20))
                a.append(b)
        Button(master=self.main_frm,text='save',command=lambda:self.save(a,path,value_inside)).pack()        
        self.l_out = tkinter.Label(master=self.main_frm,text = '')
        self.l_out.pack()    
    def save(self,a,path,value_inside):
        can_add = True
        a = [x.get() for x in a if x.get()!='']
        if len(a) == len(self.df.columns):
            index = self.df.index.name
            for index,val in enumerate(dict(zip(self.df.columns,a))):
                    if self.df[val].dtype !='O':
                        if a[index].isnumeric():
                            a[index] = np.int64(a[index])
                        else:
                            can_add = False
            if can_add:
                self.df = self.df.append(dict(zip(self.df.columns,a)),ignore_index=True)
                if index != None:
                    self.df.index.name = index
                    self.df.to_csv(path,index=True,index_label=index)
                else:
                    self.df = self.df.append(dict(zip(self.df.columns,a)),ignore_index=True)
                    self.df.reset_index(drop=True, inplace=True)
                    self.df.to_csv(path,index=False)
            if value_inside == 'Students':
                self.students = self.df
            elif  value_inside == 'Lecturers':
                self.lecturers = self.df
            elif  value_inside == 'Courses':
                self.courses = self.df
            else:
                self.take= self.df
            if can_add:
                self.l_out.config(fg='blue' ,text='Saved!',font=('Helvetica', 10, 'bold'))
            else:
                self.l_out.config(fg='red' ,text='Invalid input!',font=('Helvetica', 10, 'bold'))
        else:
            self.l_out.config(fg='red' ,text='Please, Fill all the entries',font=('Helvetica', 10, 'bold'))
        
        
    def title(self,name):  
        self.title_name.config(text=name.upper(),font=("Courier", 44))
    
    def search_page(self):
        self.root.geometry("1000x1000")
        self.clear()
        self.title('SHOW TABLES')
        self._make_main_frame()
        
        value_search = tkinter.StringVar(self.main_frm)
        value_search.set("Select an Table")
        
        question_menu = tkinter.OptionMenu(self.main_frm, value_search, *Program.search_list,command= self.build)
        question_menu.pack()

        query_val = tkinter.StringVar(self.main_frm)
        query_val.set("Select an Query")
        self.queries_list = ['Show the best student','Show students older then','Show number of failed students per course',
                             'Show a table of average grade per course','Show a table of average grade per student',
                             'show information of student by ID','Show a table of the students of the most senior lecturer','Show the weekly hours for every student',
                             'Show the students that study more than X NAZ']
        queries_menu = tkinter.OptionMenu(self.main_frm, query_val, *self.queries_list,command= self.queries_func)
        queries_menu.pack()
        self.queries_func(query_val)
        self.back_frame = Frame(master=self.root).pack()
        self.back = Button(master=self.back_frame,text="Back",command=self.menu_page).pack(pady=10)
        self.bn_cancel = Button(master=self.back_frame,text="Exit",command=self.root.destroy).pack(pady=10)
        self.main()
        
    def build(self,value_search):
        list = self.main_frm.winfo_children()
        self.clean(2)
        self.build_table(value_search) 
            
    def clean(self,n):
        list = self.main_frm.winfo_children()
        if len(list)> n: 
            for l in list[n:]:
                l.destroy()
    def queries_func(self,query_val):
        list = self.main_frm.winfo_children()
        self.clean(2)
        if self.queries_list[0] == query_val:
            best_student = self.students.loc[self.take.groupby("Student ID").mean()[self.take.groupby("Student ID").mean()['Grade'] == self.take.groupby("Student ID").mean()['Grade'].max()].Grade.index ]
            best_student['Grade'] = self.take.groupby("Student ID").mean().max().Grade
            self.build_query(best_student)
            
        elif(self.queries_list[1] == query_val):
            self._make_label("Please enter age X greater")
            e1 = self._make_entry()
            button = Button(self.main_frm,text="Enter",command=lambda: self.age(e1.get())).pack()
            
        elif(self.queries_list[2] == query_val):
            failed = self.courses.loc[self.take[self.take.Grade < 60].groupby('Course ID').count().index]
            failed["Number of failed students"] = self.take[self.take.Grade < 60].groupby('Course ID').count()['Student ID']
            self.build_query(failed[['Name','Number of failed students']])    
            self.graph_bar(x=failed['Name'],y=failed['Number of failed students'])
        
        
        elif(self.queries_list[3] == query_val):
            table = self.courses.join(self.take.groupby('Course ID').mean().Grade)[['Name', 'NAZ', 'Grade']]
            self.build_query(table)
            self.graph_bar(x=table['Name'],y=table['Grade'])
            
        elif(self.queries_list[4] == query_val):
            table = self.students.join(self.take.groupby('Student ID').mean().Grade)
            self.build_query(table)
            self.graph_bar(x=table['First Name']+"\n"+table['Last Name'],y=table.Grade)
            
        elif(self.queries_list[6] == query_val):
            self.courses.join(self.take.groupby('Course ID').mean().Grade)[['Name', 'NAZ', 'Grade']]
            self.build_query(self.lecturers[self.lecturers.Seniority == max(self.lecturers.Seniority)])
            
        
        elif(self.queries_list[5] == query_val):
            self._make_label("Please enter ID to the show info")
            e1 = self._make_entry()
            self.l_out = tkinter.Label(master=self.main_frm,text='')
            Button(self.main_frm,text="Enter",command=lambda: self.show_course_by_ID(e1.get())).pack()
            self.l_out.pack()
            
        elif(self.queries_list[7] == query_val):
            table = self.take.join(self.courses['Weekly Hours'],'Course ID').groupby('Student ID').sum().join(self.students[['First Name','Last Name']],'Student ID')[['First Name','Last Name','Weekly Hours']]
            self.build_query(table)
            self.graph_bar(x=table['First Name']+"\n"+table['Last Name'],y=table['Weekly Hours'])
        elif(self.queries_list[8] == query_val):
            self._make_label("Please enter NAZ")
            e1 = self._make_entry()
            button = Button(self.main_frm,text="Enter",command=lambda: self.show_student_by_naz(e1.get())).pack()
    def show_course_by_ID(self,x):
        self.clean(6)
        if x.isnumeric():
            q = self.courses.loc[self.take[self.take['Student ID'] == int(x)]['Course ID']]
            if not q.empty:
                df = self.students.loc[int(x)]
                self.l_out.config(text=f"Information: \nName:{df['First Name']} {df['Last Name']}, School: {df['School']}",font=('Helvetica', 14, 'bold'))
            self.build_query(q)
        else:
            self.l_out.config(text='Invalid input!!',font=('Helvetica', 10, 'bold'))
            
    def graph_bar(self,x,y):
        figure = sns.barplot(y=y,x=x).figure
        plt.Figure(figsize=(15,20))
        canvas = FigureCanvasTkAgg(figure, master=self.main_frm)
        canvas.draw()
        canvas.get_tk_widget().pack(fill=BOTH,expand=True)
    
    def age(self,x):
        self.clean(5)
        if x.isnumeric():
            self.build_query(self.students[self.students['Age'] > int(x)])
        else:
            Label(master=self.main_frm,text='Invalid input!!!!!!!',font=('Helvetica', 10, 'bold')).pack()
   
    def show_student_by_naz(self,x):
        self.clean(5)
        if x.isnumeric():
            table = self.take.join(self.courses['NAZ'],'Course ID').groupby('Student ID').sum().join(self.students[['First Name','Last Name']],
                                                                                                 'Student ID')[[ 'First Name', 'Last Name', 'NAZ']][self.take.join(self.courses['NAZ'],'Course ID').groupby('Student ID').sum().join(self.students[['First Name','Last Name']],
                                                                                                 'Student ID')[[ 'First Name', 'Last Name', 'NAZ']]['NAZ'] > int(x)]
            if table.empty == False:            
                self.build_query(table)
                self.graph_bar(x=table['First Name']+"\n"+table['Last Name'],y=table['NAZ'])
            else:
                self._make_label('NO STUDENTS FOUND')

        else:
            self._make_label('Invalid input!').config(fg='red',font=('Helvetica', 10, 'bold'))
        
    def build_query(self,df):
        if df.empty == False:
            f = Frame(self.main_frm)
            f.pack(fill=BOTH,expand=True)
            self.table = pt = Table(f, dataframe=df,editable=False,)
            pt.autoResizeColumns()
            pt.show()
        else:
            self._make_label('NO STUDENTS FOUND')
    
    def build_table(self,value_search):
        f = Frame(self.main_frm)
        f.pack(fill=BOTH,expand=True)
        if value_search == 'Students':
            df = self.students
        elif  value_search == 'Lecturers':
            df = self.lecturers
        elif  value_search == 'Courses':
            df = self.courses
        else:
            df = self.take
        self.table = pt = Table(f, dataframe=df,editable=False)
        pt.show()
        
    def clear(self):
        list = self.root.winfo_children()
        for l in list[1:]:
            l.destroy()

    def main(self):
        self.root.mainloop()
```
